```python
from IPython.utils.process import system
from typing_extensions import clear_overloads
from IPython.core.display import clear_output

          #TallerNo4 - Funciones 

#Función mostrar datos de contacto
def DContacto():
  #Mostrar datos de contacto
  print("IDENTIFICACIÓN")
  print("   Tipo de documento: " + TipoDocumento)
  print("   Número de identificación: " + NoIdentidad)
  print("NOMBRES Y APELLIDOS")
  print("   Nombres: " + Nombres)
  print("   Apellidos: " + Apellidos)
  print("DATOS DE CONTACTO")
  print("   Dirección: " + Direccion)
  print("   Teléfono: " + Telefono)

#Función mostrar datos personales
def DPersonales():
  #Mostrar datos personales
  print("DATOS PERSONALES ")
  print("   Edad: {}".format(Edad))
  print("   Estatura: {}".format(Estatura))
  print("   Estado civil: " + EstadoCivil)
  print("   Número de Hijos: {}".format(NoHijos))

#Función mostrar datos laborales
def DLaborales():
  #Mostrar datos laborales
  print("INFORMACION LABORAL")
  print("   Sueldo: {}".format(Sueldo))
  print("   Dias laborados: {}".format(DiasLaborados))
  slash = "/" #Variable para colocar el simbolo / de separación
  print("   Su fecha de contatación es: " +DiaContrato+slash+MesContrato+slash+AñoContrato)

#Funciónn mostrar bono

          #TallerNo2 - Sentencias Condicionales

def bono():
  #Bono Pre-Pensión
  if Edad > 50:
    BonoPrePension = (Sueldo*0.05) + Sueldo
    print("Felicidades su bono de Pre-Pensión es = ${:,.0f}".format(BonoPrePension))

  #Paseo de diciembre
  if EstadoCivil == "Casado" and NoHijos >= 1:
    print("Felicidades! Tiene derecho a un viaje todo pago para Diciembre!")

  #Comisión sobre el sueldo
  if Sueldo >= 1000000 and Sueldo <=1500000:
    ComisionSueldo = Sueldo*0.02
    print("Felicidades! Su comision es = ${:,.0f}".format(ComisionSueldo)) 

  elif Sueldo >=1500001 and Sueldo <= 2000000:
    ComisionSueldo = Sueldo*0.05
    print("Felicidades! Su comisión es = ${:,.0f}".format(ComisionSueldo))

  else:
    print("Su salario no tiene comisón")
  #Bono de alimentación
  if DiasLaborados > 20 and Sueldo < 1000000:
    print("Felicidades! Tiene derecho a un bono de alimentación")

          #TallerNo3 - Ciclos repetitivos

NoEmpleados = int(input("Digite el número de empleados a registrar: "))
while NoEmpleados > 0:
  NoEmpleados = NoEmpleados - 1

          #TallerNo1 - Variables y comentarios

  #Solicita tipo y numero de documento
  print("Digite su tipo y número de documento")
  TipoDocumento = input("   Tipo de documento: ")
  NoIdentidad = input("   No de Identificación: ")

  #Solicita nombres y apellidos
  print("Digite sus nombres y apellidos")
  Nombres = input("   Nombres: ")
  Apellidos = input("   Apellidos: ")

  #Solicita dirección y teleéfono
  print("Digite su dirección y teléfono")
  Direccion = input("   Dirección: ")
  Telefono = input("   Teléfono: ")
 
  #Solicita edad y estatura
  print("Digite su edad y estatura")
  Edad = int(input("   Edad: "))
  Estatura = float(input("   Estatura: "))

  #Solicita estado civil y número de hijos
  print("Digite su estado civil y número de hijos")
  EstadoCivil = input("   Estado Civil: ")
  NoHijos = int(input("   Numero de hijos: "))

  #Solicita Fecha de contratación
  print("Digite su fecha de contratación")
  DiaContrato = input("   Día: ")
  MesContrato = input("   Mes: ")
  AñoContrato = input("   Año: ")

  #Solicita sueldo y dias laborados
  print("Digite su sueldo y días laborados")
  Sueldo = float(input("   Sueldo: "))
  DiasLaborados = int(input("   Dias laborados: "))
  
          #TallerNo4 - Funciones 

  #Mostrar datos invocando funciones
  print("\n")
  #Muestra bonificación
  MostrarBono = int(input("Digite 1 para consultar bonificacion o 0 para continuar: "))
  while MostrarBono > 0:
    bono()
    MostrarBono = MostrarBono - 1
  print("\n")
  #Muestra datos de contacto
  MostrarDC = int(input("Digite 1 para consultar datos de contacto o 0 para continuar: "))
  while MostrarDC > 0:
    DContacto()
    MostrarDC = MostrarDC - 1
  print("\n")
  #Muestra datos personales
  MostrarDP = int(input("Digite 1 para consultar datos personales o 0 para continuar: "))
  while MostrarDP > 0:
    DPersonales()
    MostrarDP = MostrarDP - 1
  print("\n")
  #Muestra datos laborales
  MostrarDL = int(input("Digite 1 para consultar datos laborales o 0 para continuar: "))
  while MostrarDL > 0:
    DLaborales()
    MostrarDL = MostrarDL - 1
```
